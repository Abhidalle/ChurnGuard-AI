```python
import pandas as pd

# 1. INPUT CONFIGURATION
# Set the direct path to the Cornell University eBird text file
RAW_INPUT = '/content/ebd_NP_smp_relFeb-2026.txt'
CSV_OUTPUT = 'nepal_birds.csv'

print(f"🔄 Initializing conversion of {RAW_INPUT}...")
```

```python
# 2. FILE CONVERSION
# eBird datasets are Tab-Separated (TSV); we use '\t' to parse them
# on_bad_lines='skip' ensures the process doesn't stop if a row is corrupted
df = pd.read_csv(
    RAW_INPUT,
    sep='\t',
    encoding='utf-8',
    on_bad_lines='skip',
    low_memory=False
)
```

```python
# 3. EXPORT TO CSV
# index=False prevents pandas from adding an extra column of numbers to the left
df.to_csv(CSV_OUTPUT, index=False)
```

```python
# 4. DATA VERIFICATION
print("-" * 30)
print(f"✅ CONVERSION SUCCESSFUL")
print(f"📊 Dataset Stats: {len(df):,} rows and {len(df.columns)} columns")
print(f"📁 Local File Created: {CSV_OUTPUT}")
print("-" * 30)

# Display the first 5 rows to verify headers and data alignment
print(df.head())
```
